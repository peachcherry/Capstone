```python
import pandas as pd
products = pd.read_excel('/content/drive/Shareddrives/종합설계/product.xlsx')
member = pd.read_excel('/content/drive/Shareddrives/종합설계/member_convert.xlsx')
#products['sentiment_percentage'] = products['sentiment_percentage'].apply(pd.to_numeric)
matched= pd.DataFrame()
m = products['sex'] == '남성'
fm = products['sex'] == '여성'
male_products = products[m]
female_products = products[fm]
```

```python
for i in range(len(member)):
  if member.iloc[i,4] == '남성':
    h = (member.iloc[i,5]*0.99 <= male_products['Height']) & (member.iloc[i,5]*1.01 >= male_products['Height']) &(member.iloc[i,6]-3 <= male_products['Weight']) & (member.iloc[i,6]+3 >= male_products['Weight'])
    sub_h = male_products[h]
    bc = member.iloc[i,8] == sub_h['code']
    subset=sub_h[bc]

    df_sorted_by_values = subset.sort_values(by='sentiment_percentage' ,ascending=False)
    df_sorted_by_values['user_id'] = member.iloc[i,0]
    df = df_sorted_by_values[['user_id','product_id','product_name','sentiment','sex', 'Height','Weight','rating','comment','Size','url']]
   # df = pd.concat([matched,df])
    matched.append(df,sort=False,ignore_index=True)


  elif  member.iloc[i,4] == '여성':
    h = (member.iloc[i,5]*0.99 <= female_products['Height']) & (member.iloc[i,5]*1.01 >= female_products['Height']) &(member.iloc[i,6]-3 <= female_products['Weight']) & (member.iloc[i,6]+3 >= female_products['Weight'])
    sub_h = female_products[h]
    bc = member.iloc[i,8] == sub_h['code']
    subset=sub_h[bc]
    df_sorted_by_values = subset.sort_values(by='sentiment_percentage' ,ascending=False)
    df_sorted_by_values['user_id'] = member.iloc[i,0]
    df = df_sorted_by_values[['user_id','product_id','product_name','sentiment','sex', 'Height','Weight','rating','comment','Size','url']]
    df = pd.concat([matched,df])
    matched.append(df,sort=False,ignore_index=True)
```
