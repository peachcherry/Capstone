```python
#완성본
import requests
from bs4 import BeautifulSoup
from csv import writer
import pandas as pd
import openpyxl

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"}
url = f'https://www.musinsa.com/app/reviews/lists?type=&year_date=2022&month_date=&day_date=&max_rt=2022&min_rt=2009&brand=&page={n}'
res = requests.get(url, headers=headers)
soup = BeautifulSoup(res.text)
soup2 = BeautifulSoup(soup.prettify(), "html.parser")
product_lists = []
for n in range (1,1000):
  for e in soup2.select('div.review-list'):
    try:
      ss = e.select_one('.review-profile__body_information').text.strip()
    except:
      ss = ''
    product_name = e.select_one('.review-goods-information__name').text.strip()
    rating = e.select_one('span[style]')
    rating = rating['style']
    comment = e.select_one('.review-contents__text').text.strip()
    try:
      eval = e.select_one('ul.review-evaluation__list').get_text().split()
    except:
      eval = ''
    src = e.select_one('li img')
    try:
      src = src['src']
    except:
      ''
    allinfo = [product_name, ss, rating, comment, eval, src]
    product_lists.append(allinfo)

df = pd.DataFrame(product_lists)
df.to_excel('mss.xlsx', index = False, header = ['product_name', 'ss', 'rating', 'comment', 'eval', 'src'])
```

```python
#수정4-성별,키,몸무게 있는것만 && 사진있는것만 && 리뷰 붙여서 추출 && 성별, 키, 몸무게 스플릿 && 사이즈, 밝기, 색감, 두께감 스플릿 && 별점 숫자만
import requests
from bs4 import BeautifulSoup
from csv import writer
import pandas as pd
import openpyxl
import re

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"}
product_lists = []
for n in range (1,10):
  url = f'https://www.musinsa.com/app/reviews/lists?type=&year_date=2022&month_date=&day_date=&max_rt=2022&min_rt=2009&brand=&page={n}'
  res = requests.get(url, headers=headers)
  soup = BeautifulSoup(res.text)
  soup2 = BeautifulSoup(soup.prettify(), "html.parser")
  for e in soup2.select('div.review-list'):
    try:
      ss = e.select_one('.review-profile__body_information').text.strip()
      ss = ss.split(',')
      try:
        sex = ss[0]
        Height = ss[1]
        Weight = ss[2]
        product_name = e.select_one('.review-goods-information__name').text.strip()
        rating = e.select_one('span[style]')
        rating = rating['style']
        rating = re.sub(r'[^0-9]', '', rating)
        comment = e.select_one('.review-contents__text').text.strip()
        comment = ' '.join(comment.split())
        try:
          eval = e.select_one('ul.review-evaluation__list').get_text().split()
        except:
          eval = ''
        Size = eval[1]
        Bright = eval[3]
        Color = eval[5]
        Thick = eval[7]
        src = e.select_one('li img')
        try:
          src = src['src']
          allinfo = [product_name, sex, Height, Weight, rating, comment, Size, Bright, Color, Thick, src]
          product_lists.append(allinfo)
        except:
          ''

      except:
        sex = ''


    except:
      ss = ''


df = pd.DataFrame(product_lists)
df.to_excel('mss.xlsx', index = False, header = ['product_name', 'sex', 'Height', 'Weight', 'rating', 'comment', 'Size', 'Bright', 'Color', 'Thick', 'src'])
```
